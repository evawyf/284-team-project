```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
def kmeans(data):
    #split each group up according to the population color to see if it was colored correctly.
    #Legend: [EUR, AFR, EAS, AMR, SAS]
    eur = ["CEU", "TSI", "FIN", "GBR", "IBS"]
    afr = ["YRI", "LWK", "GWD", "MSL", "ESN", "ASW", "ACB"]
    eas = ["CHB", "JPT", "CHS", "CDX", "KHV"]
    amr = ["MXL", "PUR", "CLM", "PEL"]
    sas = ["GIH", "PJL", "BEB", "STU", "ITU"]

    #Perform kmeans clustering into 5 clusters
    km = KMeans(n_clusters=5).fit(data)
    
    plt.scatter(data[:,0], data[:,1], c=km.labels_, cmap='rainbow')
    plt.scatter(km.cluster_centers_[:,0] ,km.cluster_centers_[:,1], color='black')
    
    group_one = []
    group_two = []
    group_three = []
    group_four = []
    group_five = []
    group_six = []
    for i,c in enumerate(km.labels_):
        if c == 0:
            group_one += [i]
        elif c == 1:
            group_two += [i]
        elif c == 2:
            group_three += [i]
        elif c == 3:
            group_four += [i]
        elif c == 4:
            group_five += [i]
        elif c == 5:
            group_six += [i]
        else:
            print("This should never print")

    pop = []
    with open("df_pop.csv", "r") as f:
        line = f.readline()
        while line:
            line = f.readline()
            parse = line.strip().split(",")
            if len(parse) == 2:
                pop += [parse[1]]
    
    #group_one
    group_one_count = [0, 0, 0, 0, 0]
    for i in group_one:
        if pop[i] in eur:
            group_one_count[0] += 1
        elif pop[i] in afr:
            group_one_count[1] += 1
        elif pop[i] in eas:
            group_one_count[2] += 1
        elif pop[i] in amr:
            group_one_count[3] += 1
        elif pop[i] in sas:
            group_one_count[4] += 1
        else:
            print(pop[i])

    print("Group 1. Legend Index:")
    print(group_one_count.index(max(group_one_count)))
    print("Accuracy:")
    one_accuracy = max(group_one_count)/float(sum(group_one_count))
    print(one_accuracy)
    

    #group_two
    group_two_count = [0, 0, 0, 0, 0]
    for i in group_two:
        if pop[i] in eur:
            group_two_count[0] += 1
        elif pop[i] in afr:
            group_two_count[1] += 1
        elif pop[i] in eas:
            group_two_count[2] += 1
        elif pop[i] in amr:
            group_two_count[3] += 1
        elif pop[i] in sas:
            group_two_count[4] += 1
        else:
            print(pop[i])

    print("Group 2. Legend Index:")
    print(group_two_count.index(max(group_two_count)))
    print("Accuracy:")
    two_accuracy = max(group_two_count)/float(sum(group_two_count))
    print(two_accuracy)
    
    #group_three
    group_three_count = [0, 0, 0, 0, 0]
    for i in group_three:
        if pop[i] in eur:
            group_three_count[0] += 1
        elif pop[i] in afr:
            group_three_count[1] += 1
        elif pop[i] in eas:
            group_three_count[2] += 1
        elif pop[i] in amr:
            group_three_count[3] += 1
        elif pop[i] in sas:
            group_three_count[4] += 1
        else:
            print(pop[i])

    print("Group 3. Legend Index:")
    print(group_three_count.index(max(group_three_count)))
    print("Accuracy:")
    three_accuracy = max(group_three_count)/float(sum(group_three_count))
    print(three_accuracy)
    
    #group_four
    group_four_count = [0, 0, 0, 0, 0]
    for i in group_four:
        if pop[i] in eur:
            group_four_count[0] += 1
        elif pop[i] in afr:
            group_four_count[1] += 1
        elif pop[i] in eas:
            group_four_count[2] += 1
        elif pop[i] in amr:
            group_four_count[3] += 1
        elif pop[i] in sas:
            group_four_count[4] += 1
        else:
            print(pop[i])

    print("Group 4. Legend Index:")
    print(group_four_count.index(max(group_four_count)))
    print("Accuracy:")
    four_accuracy = max(group_four_count)/float(sum(group_four_count))
    print(four_accuracy)
    
    #group_five
    group_five_count = [0, 0, 0, 0, 0]
    for i in group_five:
        if pop[i] in eur:
            group_five_count[0] += 1
        elif pop[i] in afr:
            group_five_count[1] += 1
        elif pop[i] in eas:
            group_five_count[2] += 1
        elif pop[i] in amr:
            group_five_count[3] += 1
        elif pop[i] in sas:
            group_five_count[4] += 1
        else:
            print(pop[i])

    print("Group 5. Legend Index:")
    print(group_five_count.index(max(group_five_count)))
    print("Accuracy:")
    five_accuracy = max(group_five_count)/float(sum(group_five_count))
    print(five_accuracy)
    
    print("Total Accuracy:")
    total_accuracy = (one_accuracy + two_accuracy + three_accuracy + four_accuracy + five_accuracy)/5.0
    print(total_accuracy)
```
