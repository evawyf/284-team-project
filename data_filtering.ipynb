```python
%%bash

plink --vcf ~/284_proj/chr16.g5mac3.recode.vcf --extract plink.prune.in --make-bed --out ~/284_proj/chr16.ldprune
```

```python
%%bash

plink --bfile ~/284_proj/chr16.ldprune --recode vcf --out ~/284_proj/chr16.ldprune.out
```

```python
def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('##')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        dtype={'#CHROM': str, 'POS': int, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str},
        sep='\t'
    ).rename(columns={'#CHROM': 'CHROM'})
```

```python
ASSOCFILE = os.path.join(os.environ["HOME"], "284_proj", "chr16.ldprune.out.vcf")
vcf_df = read_vcf(ASSOCFILE)
```
