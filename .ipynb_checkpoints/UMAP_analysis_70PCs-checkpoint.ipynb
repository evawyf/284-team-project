```python
pop_to_color = {
    # EAS - yellow & orange
    "CHB": "yellow", "JPT": "khaki", "CHS": "gold", "CDX": "orange", "KHV": "darkorange", 
    # EUR - blue
    "CEU": "blue", "TSI": "dodgerblue", "FIN": "deepskyblue", "GBR": "darkblue", "IBS": "royalblue",
    # AFR - green
    "YRI": "green", "LWK": "forestgreen", "GWD": "limegreen", "MSL": "lime", "ESN": "greenyellow", 
    "ASW": "springgreen", "ACB": "seagreen", 
    # AMR - purple
    "MXL": "purple", "PUR": "darkviolet", "CLM": "magenta", "PEL": "deeppink", 
    # SAS - red
    "GIH": "red", "PJL": "crimson", "BEB": "firebrick", "STU": "lightcoral", "ITU": "indianred", 
    # None
    "None": "lightgrey"  
}
```

```python
sample_to_pop = {}
with open("./df_pop.csv", "r") as f:
    for line in f:
        sample, pop = line.strip().split(",")
        sample_to_pop[sample] = pop
```

```python
if not os.path.exists('./df_pop.csv'):
    samples = []
else: samples = [item.strip() for item in open('./df_pop.csv', "r").readlines()[1:]]
colors = []
for sample in samples:
    sample_population = sample.split(',')[1]
    #pop = sample_to_pop.get(sample_population, "None")
    color = pop_to_color[sample_population]
    colors.append(color)
```

## Load the top 2 PCs

```python
pc1 = pca_projected_mat[:,0]
pc2 = pca_projected_mat[:,1]
```

```python
# Make the plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(pc1, pc2, color=colors, s=5)
# Now make it look a little prettier since I can't stand matplotlib defaults...
ax.set_xlabel("PC 1", size=15)
ax.set_ylabel("PC 2", size=15)
ax.set_xticklabels(ax.get_xticks(), size=12)
ax.set_yticklabels(ax.get_yticks(), size=12)
ax.spines["top"].set_visible(False);
ax.spines["right"].set_visible(False);
ax.get_xaxis().tick_bottom();
ax.get_yaxis().tick_left();
```

```python
fig.savefig('./img/PCA_top2.png')
```

## Look at only Europeans (EUR)

```python
european_family = ["CEU","TSI", "FIN", "GBR", "IBS"]
```

```python
populations_information = open('./df_pop.csv').readlines()[1:]
population_mapping = [i.strip('\n').split(',') for i in populations_information]
```

```python
european_idxs = []
european_samples = []
for i,item  in enumerate(population_mapping):
    if item[1] in european_family:
        european_idxs.append(i)
        european_samples.append(item)
```

### Run umap on only Europeans on 70PCs

```python
european_people_pca_top100 = pca_projected_mat[european_idxs]
european_people_pca_top100.shape
```

```python
umap_european_top70PC = reducer.fit_transform(european_people_pca_top100[:,0:70])
umap_european_top70PC.shape
```

```python
pop_to_color_europeans = {'CEU':'yellow',      # Utah Residents (CEPH) with Northern and Western European Ancestry
                          'TSI':'green',       # Toscani in Italia 
                          'FIN':'red',         # Finnish in Finland
                          'GBR':'purple',      # British in England and Scotland
                          'IBS':'deepskyblue'} # Iberian Population in Spain

colors_europeans = []
for sample in european_samples:
    colors_europeans.append(pop_to_color_europeans[sample[1]])
    
```

```python
# Make the plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(umap_european_top70PC[:,0],umap_european_top70PC[:,1], color=colors_europeans, s=5)
# Now make it look a little prettier since I can't stand matplotlib defaults...
plt.title("European (EUR) - 70Cs", size=15)
ax.set_xlabel("Umap 1", size=15)
ax.set_ylabel("Umap 2", size=15)
ax.set_xticklabels(ax.get_xticks(), size=12)
ax.set_yticklabels(ax.get_yticks(), size=12)
ax.spines["top"].set_visible(False);
ax.spines["right"].set_visible(False);
ax.get_xaxis().tick_bottom();
ax.get_yaxis().tick_left();
```

```python
fig.savefig('./img/umap_EUR_on_top70_PCs.png')
```
